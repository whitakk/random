```python
# %matplotlib inline

import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup, Comment
import time

pd.options.display.max_rows = 200
```

Pull player registrar from baseball-reference:

```python
#%%time

chrs = 'abcdefghijklmnopqrstuvwyz'
players = []

for i in chrs:
    link = 'https://www.baseball-reference.com/players/'+i+'/'
    
    r = requests.get(link).content

    soup = BeautifulSoup(r,'lxml')

    ps = soup.findAll('p')
    
    for p in ps[10:-20]:
        players.append(p.text)
        
    time.sleep(5)
        
print(len(players))
```

```python
players_df = pd.DataFrame(players, columns=['Text'])

players_df['Player'] = players_df['Text'].apply(lambda x: x.split('(')[0].strip().strip('+'))
players_df['Debut'] = players_df['Text'].apply(lambda x: x.split('(')[1].split('-')[0])
players_df['End'] = players_df['Text'].apply(lambda x: x.split('(')[1].split('-')[1].strip(')'))

players_df = players_df[['Player','Debut','End']]
```

Summarize by year of debut:

```python
summ_df = players_df.groupby('Debut')['Player'].count().reset_index().rename(columns={'Debut':'Year', 'Player':'Debuts'})
summ_df['Cumulative_Debuts'] = summ_df['Debuts'].cumsum()
```
